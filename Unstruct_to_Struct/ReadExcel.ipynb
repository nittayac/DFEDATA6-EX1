```python
import pandas as pd
import numpy as np
import io
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_excel(io.BytesIO(uploaded['bio.xlsx']))
df.head(20)
```

```python
#removing N/A values from data frame
df = df.fillna("")
df.head(20)
```

```python
#remove all the blank rows
df1 = df.dropna(how='all')

#in col :14 >> find the words that ='%sent%'  Ex. present, absent
df1['Unnamed: 14'].str.contains('sent')
df1 = df1[df1['Unnamed: 14'].str.contains('sent')] 
df1
```

```python
#Exract Employess Name
df_name = df.dropna(how='all')
df_name = df_name[df_name['Unnamed: 3'].str.contains('Employee')]
df_name
```

```python
#new data fram for Present/Absent  , Punch record and Empoyess code
zippedList = list(zip(df1['Unnamed: 14'],df1['Unnamed: 15'],df_name['Unnamed: 7']))
abc = pd.DataFrame(zippedList)
abc.head()
list(abc.columns)   #show column name


#rename column name
abc.rename(columns={0:'Status'},inplace = True)
abc.rename(columns={1:'Punch Record'},inplace = True)
abc.rename(columns={2:'Empoyess code'},inplace = True)
list(abc.columns)

abc.head()
```
